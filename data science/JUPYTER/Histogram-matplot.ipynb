```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bloog_suger = [113, 85, 90, 150, 149, 88, 115, 93, 135, 77, 82, 129, 80]

plt.hist(bloog_suger)
```

```python
plt.hist(bloog_suger, bins=4, rwidth=0.56)
```

```python
plt.hist(bloog_suger, bins=[80, 100, 125, 150], rwidth=0.86, color='g')
```

```python
bloog_suger_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 120, 77, 82, 129]
bloog_suger_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([bloog_suger_men, bloog_suger_women], bins=5, rwidth=0.89, color=['b', 'r'])
```

```python
plt.hist([bloog_suger_men, bloog_suger_women], bins=[80, 100, 125, 150], rwidth=0.89, color=['g', 'orange'])
```

```python
plt.hist([bloog_suger_men, bloog_suger_women], bins=[80, 100, 125, 150], 
         rwidth=0.89, color=['g', 'orange'], label=["MEN", "WOMEN"])
plt.legend()
```

```python
plt.xlabel("Sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood Sugar Analysis")
plt.hist([bloog_suger_men, bloog_suger_women], bins=[80, 100, 125, 150], 
         rwidth=0.89, color=['g', 'orange'], label=["MEN", "WOMEN"])
plt.legend()
```

```python
plt.xlabel("Sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood Sugar Analysis")
plt.hist([bloog_suger_men, bloog_suger_women], 
         bins=[80, 100, 125, 150], 
         rwidth=0.89, color=['g', 'yellow'], 
         label=["MEN", "WOMEN"], 
         orientation='horizontal'
         )
plt.legend()
```
