```python
import pandas as pd
df = pd.read_csv('data/mausam.csv', parse_dates=['day'], dayfirst=True)
df
```

```python
df.set_index('day', inplace=True)
df
```

To fill na with 0 for all columns

```python
new_df = df.fillna(0)
new_df
```

To do forward fill with this obj.ffill() method

```python
print(f"{df}")
new_df = df.ffill() 
# df.fillna(method='ffill')
new_df
```

To do backword fill with this obj.bfill() method

```python
print(f"{df}")
new_df = df.bfill() 
# df.fillna(method='bfill')
new_df.head()
```

To fill na value with a specific values for each columns

```python
new_df = df.fillna({
   'temperature': 27,
   'windspeed': 50,
   'event':'normal',
   'Humidity': 10
})
new_df
```

Limited filling in forward fill

```python
print(f"{df}")
new_df = df.ffill(limit=1) 
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method='time')
new_df
```

```python
clean_df = df.dropna(how='all')
print(f"{clean_df}")

new_df = df.dropna()
new_df
```

Handling Missing data with Replace function

```python
my_df = pd.read_csv('data/clean_stock.csv')
my_df
```

```python
my_df.drop(columns='Unnamed: 0', inplace=True)
my_df
```

```python
import numpy as np
my_new_df = my_df.replace(-99, np.NaN)
my_new_df
```

```python
my_new_df = my_df.replace([-99, -89], np.NaN)
my_new_df
```

- I want to change NaN value only from the people column.
- so, that if there are other columns can contain NaN value not be replaced

```python
my_new_df = my_df.replace({
   'people':np.NaN
   }, 'dee')
my_new_df
```

Mapping of what you want with new values


```python
new_my_df = my_new_df.replace({
   np.NaN:0,
   '-89':1,
   '-99.0':1.1
})
new_my_df
```

Remove special charecters with regex for all columns

```python
new_df = new_my_df.replace('[A-Za-z]', '', regex=True)
new_df
```

Remove special charecters with regex at only the specified columns

```python
new_df = new_my_df.replace({'revenue':'[A-Za-z]', 'price':'[A-Za-z]'}, '', regex=True)
new_df
```

Replacing data from mapping list 

```python
score_df = pd.DataFrame({
   'score': ['exceptional', 'average', 'good', 'poor', 'average', 'exceptional'],
   'student': ['Ram', 'Maya', 'Prabhu', 'Krishna', 'Sita', 'Kabita']
})
score_df
```

Replace the non-numeric value with numeric value

```python
new_score = score_df.replace(['exceptional', 'average', 'good', 'poor'], [1, 4, 3, 2])
new_score
```
