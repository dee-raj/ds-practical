Matplotlib Tutorial 1

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [40, 57, 32, 21, 49, 19, 36]
```

```python
plt.title("weather")
plt.plot(x,y, color='green', linewidth=5, linestyle="dotted")
plt.xlabel("Day")
plt.ylabel("Temprature")
```

Matplotlib Tutorial 2

Here -- is line style, other eg:- -, --, dotted  
   r for color to be red, other eg:- g, b  
   D for dimond other eg:- +, *, 

```python
plt.plot(x, y, "--rD")
```

```python
plt.plot(x, y, color='red', linestyle='dotted', marker="D", markersize=14, mfc="#a0b366", alpha=0.6)
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_temp = [50, 52, 51, 58, 57, 59, 58]
min_temp = [43, 42, 40, 41, 33, 35, 37]
avg_temp = [46, 45, 45, 48, 45, 44, 47]
```

```python
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Weather")
plt.plot(days, max_temp, label="Max")
plt.plot(days, min_temp, label="Min")
plt.plot(days, avg_temp, label="AVg")

plt.legend(loc="lower left", fontsize='large')

plt.grid('both')
```

```python
import numpy as np
```

```python
company = ['GOOGLE', 'AMIN', 'MSFT', 'META']
revenue = [90, 112, 79, 56]
profit = [45, 60, 36, 24]

xpos = np.arange(len(company))
```

```python
plt.bar(company, revenue)
plt.title("Tech stock")
plt.ylabel("revenue in $")
```

```python
plt.bar(xpos, revenue)
plt.title("Tech stock")
plt.ylabel("revenue in $")
```

```python
plt.xticks(xpos, company)
plt.bar(xpos, revenue)
plt.title("Tech stock")
plt.ylabel("revenue in $")
```

```python
plt.title("Tech stock")
plt.ylabel("revenue in $")

plt.xticks(xpos, company)
plt.bar(xpos -0.2, revenue, width=0.4, label="Revenue")
plt.bar(xpos +0.2, profit, width=0.4, label="Profit")

plt.legend(loc='best')
```

```python
plt.title("Tech stock")
plt.ylabel("revenue in $")

plt.yticks(xpos, company)
plt.barh(xpos, revenue, label="Revenue",align='center')
plt.barh(xpos, profit, label="Profit",align='center')

plt.legend(loc='best')
```
