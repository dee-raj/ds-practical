Excel files

```python
ex_df = pd.read_excel('data/Products.xlsx', 'Sheet1')
ex_df
```

```python
def convert_na_value(cell):
   if cell =="n.a" or  cell ==" ":
      return 3.14
   return cell
excel_df = pd.read_excel("data/products.xlsx", 'Sheet1', 
   converters={
      'Qtr 4':convert_na_value,
      'Qtr 3':convert_na_value,
      'Qtr 2':convert_na_value,
      'Qtr 1':convert_na_value,
      'Customer': lambda cell: "dee" if cell == ' ' else cell
   },
   nrows=10
)
excel_df.head()
```

Write to excel

```python
excel_df.to_excel('data/new_product.xlsx', sheet_name='Dee1')
```

```python
excel_df.to_excel('data/products_data.xlsx', sheet_name='Dee1', index=False, startcol=3, startrow=2)
```

Writting two diffrent sheets in one excel

```python
with pd.ExcelWriter('data/product_stock.xlsx') as writer:
   df3.to_excel(writer, sheet_name='stock', index=False)
   excel_df.to_excel(writer, sheet_name='dee1', index=False)
```
