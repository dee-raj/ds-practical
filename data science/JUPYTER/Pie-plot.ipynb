```python
import matplotlib.pyplot as plt
%matplotlib inline

exp_values = [1400, 600, 100, 410, 250]
exp_labels = ['Home rent', 'Food', 'Phone / Internet Bill', 'Car', 'Other utilities']
```

```python
plt.pie(exp_values, labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_labels)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_labels, radius=1.4)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_labels, radius=1.5, autopct='%.2f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_labels, radius=1.5, autopct='%.3f%%', shadow=True)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_labels, radius=1.5, autopct='%.2f%%', explode=[0,0,0,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_labels, radius=1.5, autopct='%.2f%%', explode=[0, 0.1, 1, 0, 0.3])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_labels, radius=1.5, startangle=45)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_labels, radius=1.5, autopct='%.2f%%', explode=[0, 0.1, -0.3, 0, 0.3], startangle=180)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_labels, radius=1.5, autopct='%.2f%%', explode=[0, 0, 0.5, 0, 0.3])

plt.legend(loc='upper right')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_labels, radius=1.5, autopct='%.2f%%', explode=[0, 0, 0.5, 0, 0.3], startangle=90)

plt.legend(loc='upper left')
plt.show()
```
